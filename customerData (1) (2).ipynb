```python
# Mapping: Coversion of Qualitative features to Ordinal Categorical like
```

```python
education = {'HighSchool': 0, 'College': 1, 'Bachelor': 2, 'Masters': 3}
df['education'] = df['education'].map(education)
```

```python
gender = {'Female': 0, 'Male': 1}
df['gender'] = df['gender'].map(gender)
```

```python
region = {'North': 0, 'South': 1, 'East': 2, 'West':3}
df['region'] = df['region'].map(region)
```

```python
loyalty_status = {'Regular': 0, 'Silver': 1, 'Gold': 2}
df['loyalty_status'] = df['loyalty_status'].map(loyalty_status)
```

```python
purchase_frequency = {'rare': 0, 'occasional': 1, 'frequent': 2}
df['purchase_frequency'] = df['purchase_frequency'].map(purchase_frequency)
```

```python
product_category = {'Books': 0, 'Clothing': 1, 'Food': 2, 'Health': 3, 'Home': 4, 'Electronics': 5, 'Beauty': 6 }
df['product_category'] = df['product_category'].map(product_category)
```

```python
# Data Training using Descision tree and Logistics Models
```

```python
df.set_index('age')
```

```python
features = ['age', 'gender', 'education', 'region', 'loyalty_status', 'purchase_frequency', 'purchase_amount', 'product_category', 'promotion_usage']
```

```python
X = df[features]
Y = df['satisfaction_score']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state= None)
```

```python
from sklearn import tree
my_classifier = tree.DecisionTreeClassifier()
my_classifier.fit(X_train, Y_train)
```

```python
print (my_classifier.predict(X_test))
```

```python
# lOGISTICS REGRESSION
from sklearn.linear_model import LogisticRegression
```

```python
lr_model = LogisticRegression()
lr_model.fit(X_train, Y_train)
```

```python
print (lr_model.predict(X_test))
```

```python
# Compare Basic Performance

dt_accuracy = my_classifier.score(X_test, Y_test)
lr_accuracy = lr_model.score(X_test, Y_test)

print(f"Decision Tree Accuracy: {dt_accuracy}")
print(f"Logistic Regression Accuracy: {lr_accuracy}")
```
